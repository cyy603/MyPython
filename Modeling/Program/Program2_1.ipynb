```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LogisticRegression
```

Bp - 1, Hp - 0

Pre-process the data

Constructing Bp

```python
NW_Bp = pd.read_csv('../feature/NW_Bp.csv')
W_Bp = pd.read_csv('../feature/W_Bp.csv')
```

```python
NW_Bp = NW_Bp.drop(columns = 'Unnamed: 0')
W_Bp = W_Bp.drop(columns = 'Unnamed: 0')
NW_Bp = NW_Bp.drop(columns = '总含量')
W_Bp = W_Bp.drop(columns = '总含量')
NW_Bp = NW_Bp.drop(columns = '文物采样点')
W_Bp =W_Bp.drop(columns = '文物采样点')
```

```python
NW_Bp['表面风化'] = NW_Bp['表面风化'].replace(['无风化', '风化'], 0)
```

```python
NW_Bp
```

```python
from sklearn.preprocessing import LabelEncoder

encode = ['纹饰', '类型', '颜色']
for i in encode:
    le = LabelEncoder()
    NW_Bp[i] = le.fit_transform(NW_Bp[i])

NW_Bp
```

```python
W_Bp['表面风化'] = W_Bp['表面风化'].replace(['无风化', '风化'], 1)
W_Bp
```

```python
from sklearn.preprocessing import LabelEncoder

for i in encode:
    le = LabelEncoder()
    W_Bp[i] = le.fit_transform(W_Bp[i])

W_Bp
```

```python
NW_Bp
```

```python
Bp = pd.concat([NW_Bp, W_Bp], axis = 0)
Bp
```

```python
Bp['label'] = np.ones((49, 1))
Bp
```

Constructing Hp

```python
NW_Hp = pd.read_csv('../feature/NW_Hp.csv')
W_Hp = pd.read_csv('../feature/W_Hp.csv')

NW_Hp = NW_Hp.drop(columns = 'Unnamed: 0')
W_Hp = W_Hp.drop(columns = 'Unnamed: 0')
NW_Hp = NW_Hp.drop(columns = '总含量')
W_Hp = W_Hp.drop(columns = '总含量')
NW_Hp = NW_Hp.drop(columns = '文物采样点')
W_Hp =W_Hp.drop(columns = '文物采样点')

NW_Hp['表面风化'] = NW_Hp['表面风化'].replace(['无风化', '风化'], 0)

for i in encode:
    le = LabelEncoder()
    NW_Hp[i] = le.fit_transform(NW_Hp[i])

W_Hp['表面风化'] = W_Hp['表面风化'].replace(['无风化', '风化'], 1)

for i in encode:
    le = LabelEncoder()
    W_Hp[i] = le.fit_transform(W_Hp[i])


Hp = pd.concat([NW_Hp, W_Hp], axis = 0)
Hp['label'] = np.zeros((18, 1))

Hp
```

```python
Data = pd.concat([Bp, Hp], axis = 0)
Data
```

```python
train_X = Data.iloc[:, 1 : Data.shape[1] - 2]
train_y = Data.iloc[:, Data.shape[1] - 1]
```

```python
train_X
```

```python
train_y
```

```python
train_X = np.array(train_X)
train_y = np.array(train_y).ravel()
```

```python
lg = LogisticRegression(random_state = 0)
lg.fit(train_X, train_y)

pred_y = lg.predict(train_X)
```

```python
pred_y
```

```python
item = Data['文物编号']
res = np.column_stack([item, pred_y])
pd.DataFrame(res).to_excel('../Q2_answer/Q2_1/classify_rule_valid.xlsx')
```
