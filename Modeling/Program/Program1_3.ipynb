```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
mean_feature = pd.read_csv('../feature/mean feature.csv')
mean_feature
```

```python
mean_feature['Hp_change_rate'] = (mean_feature['高钾未风化'] - mean_feature['高钾风化']) / mean_feature['高钾未风化']
mean_feature.fillna(value = 0, inplace = True)
mean_feature
```

```python
mean_feature.index
```

```python
mean_feature = mean_feature.drop([14])
mean_feature
```

```python
mean_feature['Bp_change_rate'] = (mean_feature['铅钡未风化'] - mean_feature['铅钡风化']) / mean_feature['铅钡未风化']
mean_feature
```

```python
Hp_change_rate = np.array(mean_feature['Hp_change_rate']).reshape((1, 14))
Bp_change_rate = np.array(mean_feature['Bp_change_rate']).reshape((1, 14))
```

导入风化数据

```python
Data_raw = pd.read_csv('../Data/valid_data2.csv')
Data_raw
```

```python
delete = [23, 25, 29, 43, 44, 47, 53, 55, 59]
Data_raw = Data_raw.drop(delete)
Data_raw
```

```python
Data_Hp_list = [0, 2, 3, 4, 5, 6, 7, 8, 11, 12, 14, 15, 16, 17, 18, 21, 22]
Data_Bp_list = [i for i in Data_raw.index if i not in Data_Hp_list]
len(Data_Hp_list) + len(Data_Bp_list)
```

```python
Data_Hp = Data_raw.loc[Data_Hp_list]
Data_Hp
```

```python
Data_Bp = Data_raw.loc[Data_Bp_list]
Data_Bp
```

从Bp和Hp中提取风化值

```python
W_Bp = Data_Bp.loc[Data_Bp['风化'] == 1]
W_Bp = W_Bp.drop(columns = '风化')
W_Bp
```

```python
W_Hp = Data_Hp.loc[Data_Hp['风化'] == 1]
W_Hp = W_Hp.drop(columns = '风化')
W_Hp
```

```python
pd.DataFrame(Hp_change_rate).to_excel('../Q1_answer/Q1_3/Hp/Hp_change_rate.xlsx')
```

```python
pd.DataFrame(Bp_change_rate).to_excel('../Q1_answer/Q1_3/Pb/Pb_change_rate.xlsx')
```

Q1.3 result

```python
Hp_res = (W_Hp) * (1 + Hp_change_rate)
Hp_res.to_excel('../Q1_answer/Q1_3/Hp/Hp_predict_res.xlsx')
```

```python
Bp_res = W_Bp * (1 + Bp_change_rate)
Bp_res.to_excel('../Q1_answer/Q1_3/Pb/Pb_predict_res.xlsx')
```

Compare result and real value

```python
Hp_NW_feature = mean_feature['高钾未风化']
Pb_NW_feature = mean_feature['铅钡未风化']
```

```python
Hp_NW_feature.to_excel('../Q1_answer/Q1_3/Hp/Hp_NW_mean_feature.xlsx')
Pb_NW_feature.to_excel('../Q1_answer/Q1_3/Pb/Pb_NW_mean_feature.xlsx')
```

```python
Hp_res_mean = Hp_res.mean()
Hp_res_mean.to_excel('../Q1_answer/Q1_3/Hp/Hp_res_mean.xlsx')
```

```python
Pb_res_mean = Bp_res.mean()
Pb_res_mean.to_excel('../Q1_answer/Q1_3/Pb/Pb_res_mean.xlsx')
```

```python
Hp_diff = np.abs(np.array(Hp_res_mean) - np.array(Hp_NW_feature))
pd.DataFrame(Hp_diff).to_excel('../Q1_answer/Q1_3/Hp/Hp_res_diff.xlsx')
```

```python
Pb_diff = np.abs(np.array(Pb_res_mean) - np.array(Pb_NW_feature))
pd.DataFrame(Pb_diff).to_excel('../Q1_answer/Q1_3/Pb/Pb_res_diff.xlsx')
```
