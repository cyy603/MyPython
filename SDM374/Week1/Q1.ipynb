Main part of Linear Regression:

```python
def h(X, theta):

    return X @ theta.T

def cost_function(X, y, theta):

    return np.sum(np.power(h(X, theta) - y, 2)) / (2 * X.shape[0])

def gradient_descent(X, y, alpha, epoch):
    theta = np.zeros((1, 2))
    cost = np.zeros(epoch)
    for i in range(epoch):
        partial_theta = np.sum((h(X, theta) - y) * X, axis = 0) / X.shape[0]
        theta = theta - alpha * partial_theta
        cost[i] = cost_function(X, y, theta)
    return theta, cost
```

Show the original data

```python
path = '../Week1_data/data_1.csv'
data = pd.read_csv(path)

plt.scatter(data['x'], data['y'])
plt.xlabel('size')
plt.ylabel('price')
plt.title('original data')
```

```python
train_X = np.array(data['x'])
train_y = np.array(data['y'])
train_X = np.column_stack([np.ones((train_X.shape[0], 1)), train_X])
train_y = train_y.reshape(101, 1)

train_X, train_y
```

Train

```python
epoch = 1000
alpha = 0.01
theta, cost = gradient_descent(train_X, train_y, alpha, epoch)

cost[len(cost) - 1]
```

```python
y = h(train_X, theta)
y
```

Exhibit performance

```python
plt.scatter(data['x'], data['y'], label = 'original data', color = 'r')
plt.plot(train_X[:, 1], y, label = 'linear regression', color = 'green')
plt.legend()
```

Show the cost decreasing

```python
plt.plot(np.arange(epoch), cost, label = 'cost', color = 'r')
plt.xlabel('iteration')
plt.ylabel('cost')
plt.title('cost-iteration diagram')
plt.legend()
```

探讨不同步长对梯度下降的影响

```python
alpha_arr = np.linspace(0.001, 0.1, num = 20)
alpha_arr
```

```python

fig1, ax_list1 = plt.subplots(nrows = 5, ncols = 4, figsize = (30, 30))
fig2, ax_list2 = plt.subplots(nrows = 5, ncols = 4, figsize = (30, 30))
index = 0

for i in range(5):
    for j in range(4):
        temp_alpha = float(alpha_arr[index])

        temp_theta, temp_cost = gradient_descent(train_X, train_y, alpha = temp_alpha, epoch = 1000)
        temp_y = h(train_X, temp_theta)

        ax_list1[i, j].scatter(data['x'], data['y'], label = 'original data', color = 'r')
        ax_list1[i, j].plot(train_X[:, 1], temp_y, label = 'linear regression', color = 'green')
        ax_list1[i, j].set_xlabel('size')
        ax_list1[i, j].set_ylabel('price')
        ax_list1[i, j].set_title('Linear Regression with learning rate %f' % temp_alpha)
        ax_list1[i, j].legend()

        ax_list2[i, j].plot(np.arange(1000), temp_cost, label = 'cost')
        ax_list2[i, j].set_xlabel('iteration')
        ax_list2[i, j].set_ylabel('cost')
        ax_list2[i, j].set_title('Cost of Linear Regression with learning rate %f' % temp_alpha)
        ax_list2[i, j].legend()

        index += 1

plt.show()
```
